Input functions for Estimator

```python
import tensorflow as tf

def combined_dataset(features, labels):
    # Performance tip: Simple tuple (features, labels) is usually better 
    # than the dictionary format unless you specifically need named inputs.
    dataset = tf.data.Dataset.from_tensor_slices((features, labels))
    return dataset

def train_input_fn(x_train, y_train):
    dataset = combined_dataset(x_train, y_train)
    # Shuffle first, then repeat, then batch. 
    # Use AUTOTUNE for prefetching to let TF manage memory automatically.
    return dataset.shuffle(50000).repeat().batch(200).prefetch(tf.data.AUTOTUNE)

def eval_input_fn(x_test, y_test):
    dataset = combined_dataset(x_test, y_test)
    return dataset.batch(1000).prefetch(tf.data.AUTOTUNE)
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
import pickle
import os

# Paths for Kaggle
TRAIN_SET = '/kaggle/working/train_set.pickle'
TEST_SET = '/kaggle/working/test_set.pickle'

# Load data
with open(TEST_SET, 'rb') as file:
    test_set = pickle.load(file)
    x_test = test_set['x']
    y_test = test_set['y']

with open(TRAIN_SET, 'rb') as file:
    train_set = pickle.load(file)
    x_train = train_set['x']
    y_train = train_set['y']
```

```python
from tensorflow.keras.utils import to_categorical

# Convert labels to (N, 5) one-hot vectors
y_train = to_categorical(y_train, num_classes=5)
y_test = to_categorical(y_test, num_classes=5)

print("New Label Shape:", y_train.shape)
# Output should be: (441841, 5)
```
